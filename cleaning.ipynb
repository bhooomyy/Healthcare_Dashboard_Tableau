```python
import pandas as pd
import math
import numpy as np
```

```python
dataset=pd.read_csv("/Users/bhoomi/Documents/GitHub/Healthcare_Dashboard_Tableau/healthcare_dataset.csv") #extracting dataset
print(dataset) # printing dataset
```

```python
dataset['Name']=dataset['Name'].str.title() #Messay name -> Proper name 
print(dataset['Name'])
```

```python
dataset.describe() 
```

```python
print(dataset[dataset.duplicated()]) 
dataset.drop_duplicates(inplace=True) #Deleting duplicates
dataset.describe()
```

```python
dataset.isnull().sum() # Looking for NULL values in each column
```

```python
dataset['Age']=dataset['Age'].astype(int)
dataset['Room Number']=dataset['Room Number'].astype(int)
dataset['Billing Amount']=round(dataset['Billing Amount'].astype(float),2)

dataset['Date of Admission']=dataset['Date of Admission'].astype('date32[pyarrow]')
dataset['Discharge Date']=dataset['Discharge Date'].astype('date32[pyarrow]')
```

```python
dataset.info()
```

```python
dataset[dataset['Name']=="Terry Wilson"]
```

```python
match_key=['Name','Medical Condition','Date of Admission','Doctor','Hospital','Insurance Provider','Billing Amount','Room Number','Admission Type','Discharge Date']
dataset.drop_duplicates(subset=(match_key),inplace=True)
print(dataset)
```

```python
dataset.to_csv('/Users/bhoomi/Documents/GitHub/Healthcare_Dashboard_Tableau/clean_healthcareDataset.csv')
```

EDA - Overview Dashboard

```python
print(f"Total Patients: ",dataset['Name'].nunique())
print(f"Total Hospitals: ",dataset['Hospital'].nunique())
print(f"Total Bills paid: ",round(dataset['Billing Amount'].sum(),2))
print(f"Average Age of Patients: ",round(dataset['Age'].sum()/dataset['Age'].count()))
print(f"Total Insurance Companies: ",dataset['Insurance Provider'].nunique())

print(f"\nNumber of patients group by ",dataset.groupby('Blood Type')['Name'].count())

print(f"\nMinimum age & Max Age:",min(dataset['Age']),max(dataset['Age']))
age_bins=[0,17,24,44,64,100]
age_label=['Kids(age<18)','Young Adults(age>=18 and <25)','Adults(age>=25 & <45)','Middle aged(age>=45 and <65)','Retired(age>65)']
dataset['Age Distribution']=pd.cut(dataset['Age'],bins=age_bins,labels=age_label)
print("Number of patients based on their age group: ",dataset.groupby('Age Distribution',observed=True)['Age'].count())

print(f"\nNumber of Males ", (dataset['Gender']=='Male').sum())
print(f"Number of Females ",(dataset['Gender']=='Female').sum())

print(f"\nNumber of perople by ",dataset.groupby('Medical Condition',observed=True)['Name'].nunique())

room_bins=[100,199,299,399,499]
room_label=['100-200','200-300','300-400','400-500']
dataset['Room Utilization']=pd.cut(dataset['Room Number'],bins=room_bins,labels=room_label)
room_utilization_pivot=dataset.pivot_table(
    index='Room Utilization',
    columns='Admission Type',
    values='Room Number',
    aggfunc='count',
    observed=True
)
result = dataset.groupby(['Room Utilization', 'Admission Type'],observed=True)['Room Number'].count()
print(f"\n",result)

print(f"\nSum of billing amount per hospital: ")
result_billing_amt=dataset.groupby('Hospital',observed=True)['Billing Amount'].sum()
print(result_billing_amt.sort_values(ascending=False).head(10))
```

EDA - Clinical Dashboard

```python
print(f"Most common ",dataset.groupby('Medical Condition',observed=True)['Name'].count().sort_values(ascending=False).head(1))
print(f"\nAvg age of patient: ",round(dataset['Age'].sum()/dataset['Age'].count()))
print(f"\n% of Normal test result: ",round(dataset[dataset['Test Results']=='Normal']['Test Results'].count()*100)/dataset['Test Results'].count())
print(f"\n% of Abnormal test result: ",round(dataset[dataset['Test Results']=='Abnormal']['Test Results'].count()*100)/dataset['Test Results'].count())
print(f"\n% of Inconclusive test result: ",round(dataset[dataset['Test Results']=='Inconclusive']['Test Results'].count()*100)/dataset['Test Results'].count())
print(f"\nCondition by frequency: ",dataset.groupby('Medical Condition')['Name'].count())
print(f"\nMedical condition by\n",dataset.groupby(['Medical Condition','Admission Type'],observed=True)['Name'].count().unstack())
print(f"\nMedical condition Vs Medicine used\n ",dataset.groupby(['Medical Condition','Medication'])['Name'].count().unstack())
print(f"\nMedical condition vs billing amount:\n ",round(dataset.groupby('Medical Condition')['Billing Amount'].mean(),2))
print(f"\nTest Result breakdown:\n ",dataset.groupby('Test Results')['Name'].count())
```

EDA - Financial Dashboard

```python
print(f"Total bill paid: ",round(dataset['Billing Amount'].sum(),2))
print(f"\nAvg bill: ",round(dataset['Billing Amount'].mean()))
print(f"\nHighest billing",dataset.groupby('Hospital',observed=True)['Name'].count().sort_values(ascending=False).head(1))
print(f"\nTop ",dataset.groupby('Insurance Provider',observed=True)['Name'].count().sort_values(ascending=False).head(1))
print(f"\nBills paid to ",dataset.groupby('Hospital',observed=True)['Billing Amount'].sum().sort_values(ascending=False).head(1))
print(f"\nMedical condition vs Insurance provider\n",dataset.groupby(['Medical Condition','Insurance Provider'])['Name'].count().unstack())
print(f"\nHospital wise avg Billing amount: ",round(dataset.groupby('Admission Type')['Billing Amount'].mean(),2))
print(f"\nBilling by Gender\n",round(dataset.groupby('Gender',observed=True)['Billing Amount'].sum(),2))
print(f"Billing by Insurance provider: ",dataset.groupby('Insurance Provider')['Billing Amount'].sum())
```

EDA - Trends

```python
dataset['Avg Stay']=(dataset['Discharge Date']-dataset['Date of Admission']).dt.days
print(f"Avg stay: ",round(dataset['Avg Stay'].mean()))
dataset['Month of Date']=dataset['Date of Admission'].dt.month_name()
print(f"\nPeak Month: ",dataset.groupby('Month of Date',observed=True)['Name'].count().sort_values(ascending=False).head(1))
print(f"\nTotal admissions ",dataset['Name'].nunique())
print(f"\nBusiest ",dataset.groupby('Hospital')['Name'].nunique().sort_values(ascending=False).head(1))

dataset['Month-Year']=dataset['Date of Admission'].dt.strftime('%B-%Y')
print(f"\nAdmission by ",dataset.groupby('Month-Year')['Name'].count())
print(f"\nAvg stay by ",round(dataset.groupby('Medical Condition')['Avg Stay'].mean()))
print(f"\nSeasonal admission pattern\n ",dataset.groupby(['Month of Date','Admission Type'],observed=True)['Name'].count().unstack())
print(f"\nNumber of patients each year ",dataset.groupby(dataset['Date of Admission'].dt.year)['Name'].count())
def format_millions(x):
    return f"{x/1_000_000:.2f}M"
print(f"\nBilling over months of years\n ",dataset.groupby('Month-Year')['Billing Amount'].sum().apply(format_millions))
```

EDA- DR & Hospital Performace

```python
print(f"Total Doctors: ",dataset['Doctor'].nunique())
print(f"\nTop doctor by patients ",dataset.groupby('Doctor')['Name'].count().sort_values(ascending=False).head(1))
dataset['Stay']=(dataset['Discharge Date']-dataset['Date of Admission']).dt.days
print(f"\nAvg stay: ",round(dataset['Stay'].mean()))
print(f"\nAvg Bill: ",round(dataset['Billing Amount'].mean(),2))

print(f"\nTop 10 doctors by patient count: ",dataset.groupby('Doctor')['Name'].count().sort_values(ascending=False).head(10))
partial_testResultbyDoctorResult=dataset.groupby(['Doctor','Test Results'])['Name'].nunique().unstack()
partial_testResultbyDoctorResult['Total']=partial_testResultbyDoctorResult.sum(axis=1)
partial_testResultbyDoctorResult=partial_testResultbyDoctorResult.sort_values('Total',ascending=False).head(10)
print(f"\nTest Results by Doctors:\n ",partial_testResultbyDoctorResult)
print(f"\nNumber of patients by hospital: ",dataset.groupby('Hospital')['Name'].count().sort_values(ascending=False).head(50))
print(f"\nAvg bill by doctor ",round(dataset.groupby('Doctor')['Billing Amount'].mean(),2).sort_values(ascending=False))
stayVsBilling_result=dataset.groupby(['Stay'])['Billing Amount'].agg( 
    Mean_Billing='mean',
    Min_Billing='min',
    Max_Billing='max',
    Patient_Count='count').round(2)
billing_bins=[0,10000,20000,30000,40000,50000]
billing_label=['0-10k','10k-20k','20k-30k','30k-40k','40k-50k']
dataset['Billing Range']=pd.cut(dataset['Billing Amount'],bins=billing_bins,labels=billing_label)
result=dataset.groupby(['Stay','Billing Range'],observed=True)['Hospital'].count().unstack()
print(result)
```
