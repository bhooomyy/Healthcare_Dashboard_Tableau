```python
import pandas as pd
```

```python
dataset=pd.read_csv("/Users/bhoomi/Documents/GitHub/Healthcare_Dashboard_Tableau/healthcare_dataset.csv") #extracting dataset
print(dataset) # printing dataset
```

```python
dataset['Name']=dataset['Name'].str.title() #Messay name -> Proper name 
print(dataset['Name'])
```

```python
dataset.describe() 
```

```python
print(dataset[dataset.duplicated()]) 
dataset.drop_duplicates(inplace=True) #Deleting duplicates
dataset.describe()
```

```python
dataset.isnull().sum() # Looking for NULL values in each column
```

```python
dataset['Age']=dataset['Age'].astype(int)
dataset['Room Number']=dataset['Room Number'].astype(int)
dataset['Billing Amount']=round(dataset['Billing Amount'].astype(float),2)

dataset['Date of Admission']=dataset['Date of Admission'].astype('date32[pyarrow]')
dataset['Discharge Date']=dataset['Discharge Date'].astype('date32[pyarrow]')
```

```python
dataset.info()
```

```python
dataset[dataset['Name']=="Terry Wilson"]
```

```python
match_key=['Name','Medical Condition','Date of Admission','Doctor','Hospital','Insurance Provider','Billing Amount','Room Number','Admission Type','Discharge Date']
dataset.drop_duplicates(subset=(match_key),inplace=True)
print(dataset)
```

```python
dataset.to_csv('/Users/bhoomi/Documents/GitHub/Healthcare_Dashboard_Tableau/clean_healthcareDataset.csv')
```

```python
print(f"Total Patients: ",dataset['Name'].nunique())
print(f"Total Hospitals: ",dataset['Hospital'].nunique())
print(f"Total Bills paid: ",round(dataset['Billing Amount'].sum(),2))
print(f"Average Age of Patients: ",round(dataset['Age'].sum()/dataset['Age'].count()))
print(f"Total Insurance Companies: ",dataset['Insurance Provider'].nunique())

print(f"\nNumber of patients group by ",dataset.groupby('Blood Type')['Name'].count())

print(f"\nMinimum age & Max Age:",min(dataset['Age']),max(dataset['Age']))
age_bins=[0,17,24,44,64,100]
age_label=['Kids(age<18)','Young Adults(age>=18 and <25)','Adults(age>=25 & <45)','Middle aged(age>=45 and <65)','Retired(age>65)']
dataset['Age Distribution']=pd.cut(dataset['Age'],bins=age_bins,labels=age_label)
print("Number of patients based on their age group: ",dataset.groupby('Age Distribution',observed=True)['Age'].count())

print(f"\nNumber of Males ", (dataset['Gender']=='Male').sum())
print(f"Number of Females ",(dataset['Gender']=='Female').sum())

print(f"\nNumber of perople by ",dataset.groupby('Medical Condition',observed=True)['Name'].nunique())

room_bins=[100,199,299,399,499]
room_label=['100-200','200-300','300-400','400-500']
dataset['Room Utilization']=pd.cut(dataset['Room Number'],bins=room_bins,labels=room_label)
room_utilization_pivot=dataset.pivot_table(
    index='Room Utilization',
    columns='Admission Type',
    values='Room Number',
    aggfunc='count',
    observed=True
)
result = dataset.groupby(['Room Utilization', 'Admission Type'],observed=True)['Room Number'].count()
print(f"\n",result)

print(f"\nSum of billing amount per hospital: ")
result_billing_amt=dataset.groupby('Hospital',observed=True)['Billing Amount'].sum()
print(result_billing_amt.sort_values(ascending=False).head(10))
```
